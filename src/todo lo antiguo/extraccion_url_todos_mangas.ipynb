```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Configuración del webdriver
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL a la que se debe acceder
url = "https://inmanga.com/manga/consult"
driver.get(url)

# Esperar hasta que el botón sea visible
wait = WebDriverWait(driver, 10)
button = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="searchMoreButton"]')))

# Obtener el ancho inicial de la página
initial_width = driver.execute_script("return document.body.scrollWidth")

while True:
    # Hacer clic en el botón
    button.click()
    
    # Esperar un momento para que la página cargue contenido nuevo
    time.sleep(4)
    
    # Esperar hasta que el botón sea visible nuevamente antes de proceder
    try:
        button = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="searchMoreButton"]')))
    except:
        break
    
    # Obtener el ancho actual de la página
    current_width = driver.execute_script("return document.body.scrollWidth")
    
    print(initial_width)
    print(current_width)
    
    # Si el ancho de la página no cambia, salir del bucle
    if current_width == initial_width:
        break
    else:
        initial_width = current_width

# Extraer todos los href de los elementos <a> después de hacer scroll
elements = driver.find_elements(By.TAG_NAME, 'a')
hrefs = [element.get_attribute('href') for element in elements if element.get_attribute('href')]

# Imprimir los href extraídos
for href in hrefs:
    print(href)

# Cerrar el navegador
driver.quit()
```
